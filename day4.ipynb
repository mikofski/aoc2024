XMAS word search

```python
import logging
import pathlib

logging.basicConfig()
LOGGER = logging.getLogger("aoc2024:day4")
LOGGER.setLevel(logging.DEBUG)
LOGGER.debug('testing')
```

```python
TEST_DATA = """MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX"""
TEST_XMAS_CT = 18
print(TEST_DATA)
```

```python
TEST_DATA = [list(test_line) for test_line in TEST_DATA.split('\n')]
TEST_DATA
```

```python
def get_all_indices(strlist, xchr):
    start = 0
    while start < len(strlist):
        try:
            x = strlist.index(xchr, start)
        except:
            return
        else:
            start = x+1
            yield x

x = [[idx for idx in get_all_indices(test_line, "X")] for test_line in TEST_DATA]
x
```

```python
def get_search_directions(row, col, nrows):
    test_locs = {
        'back': (row, col-1),
        'front': (row, col+1)}
    diagdown = {
        'backdown': (row+1, col-1),
        'down':(row+1, col),
        'frontdown': (row+1, col+1)}
    diagup = {
        'backup': (row-1, col-1),
        'up':(row-1, col),
        'frontup': (row-1, col+1)}
    if row == 0:
        test_locs.update(diagdown)
    elif (row+1) == nrows:
        test_locs.update(diagup)
    else:
        test_locs.update(diagdown)
        test_locs.update(diagup)
    return test_locs


XMAS = 'XMAS'
NCHRS = len(XMAS)
def test_all_around(wordsearch, row, col, xchr=XMAS[1], direction=None):
    allretv = []
    retv = [(None,) * 3]
    nrows = len(wordsearch)
    ncols = len(wordsearch[0])  # it is actually square and not jagged
    test_locs = get_search_directions(row, col, nrows)
    if direction is None:
        LOGGER.debug(f'no direction searching for {xchr} around  row {row}, col {col}')
        for k, v in test_locs.items():
            if col < NCHRS-1 and k in ("back", "backdown", 'backup'):
                LOGGER.debug(f'too small for {XMAS} in {k} direction!')
                continue
            if ncols-col < NCHRS-1 and k in {"front", 'frontdown', 'frontup'}:
                LOGGER.debug(f'too small for {XMAS} in {k} direction!')
                continue
            if row < NCHRS-1 and k in ("up", "frontup", 'backup'):
                LOGGER.debug(f'too small for {XMAS} in {k} direction!')
                continue
            if nrows-row < NCHRS-1 and k in {"down", 'frontdown', 'backdown'}:
                LOGGER.debug(f'too small for {XMAS} in {k} direction!')
                continue
            LOGGER.debug(f'is {xchr} in {k} direction?')
            if wordsearch[v[0]][v[1]] == xchr:
                LOGGER.debug('yep!')
                try:
                    retv = test_all_around(wordsearch, v[0], v[1], xchr=XMAS[2], direction=k)
                except Exception as exc:
                    LOGGER.debug(exc)
                    LOGGER.debug('... trying a different direction -->')
                    continue
                else:
                    LOGGER.debug('success!')
                    retv = retv[0], (row, col), retv[1]
                    allretv.append(retv)
                    # break
            else:
                LOGGER.debug('nope!')
    else:
        k, v = direction, test_locs[direction]
        LOGGER.debug(f'searching in row {v[0]}, col {v[1]} for {xchr} in {direction} direction')
        if wordsearch[v[0]][v[1]] == xchr:
            LOGGER.debug('yep!')
            if xchr == XMAS[-1]:
                return k, (v[0], v[1])
            retv = test_all_around(wordsearch, v[0], v[1], xchr=XMAS[3], direction=k)
        else:
            raise Exception(f'{xchr} not found at ({v[0]}, {v[1]})')
    if not allretv:
        return retv
    return allretv

test_all_around(TEST_DATA, 0, col=x[0][0])
```

```python
xmas_found = [test_all_around(TEST_DATA, n, col=xxx) for n, xx in enumerate(x) for xxx in xx]
xmas_found
```

```python
num_xmas_found = len([x for y in xmas_found for x in y if x != (None,)*3])
print(num_xmas_found)
assert num_xmas_found == TEST_XMAS_CT
```
