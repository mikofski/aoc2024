XMAS word search

```python
test_data = """MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX"""
test_xmas_ct = 18
print(test_data)
```

```python
test_data = [list(test_line) for test_line in test_data.split('\n')]
test_data
```

```python
def get_all_indices(strlist, xchr):
    start = 0
    while start < len(strlist):
        try:
            x = strlist.index(xchr, start)
        except:
            return
        else:
            start = x+1
            yield x

x = [[idx for idx in get_all_indices(test_line, "X")] for test_line in test_data]
x
```

```python
def get_search_directions(row, col, nrows):
    test_locs = {
        'back': (row, col-1),
        'front': (row, col+1)}
    diagdown = {
        'backdown': (row+1, col-1),
        'down':(row+1, col),
        'frontdown': (row+1, col+1)}
    diagup = {
        'backup': (row-1, col-1),
        'up':(row-1, col),
        'frontup': (row-1, col+1)}
    if row == 0:
        test_locs.update(diagdown)
    elif (row+1) == nrows:
        test_locs.update(diagup)
    else:
        test_locs.update(diagdown)
        test_locs.update(diagup)
    return test_locs


XMAS = 'XMAS'
NCHRS = len(XMAS)
def test_all_around(wordsearch, row, col, xchr=XMAS[1], direction=None):
    retv = (None,) * 3
    nrows = len(wordsearch)
    ncols = len(wordsearch[0])  # it is actually square and not jagged
    test_locs = get_search_directions(row, col, nrows)
    if direction is None:
        print(f'no direction searching for {xchr} around  row {row}, col {col}')
        for k, v in test_locs.items():
            if col < NCHRS-1 and k in ("back", "backdown", 'backup'):
                print(f'too small for {XMAS} in {k} direction!')
                continue
            if ncols-col < NCHRS-1 and k in {"front", 'frontdown', 'frontup'}:
                print(f'too small for {XMAS} in {k} direction!')
                continue
            if row < NCHRS-1 and k in ("up", "frontup", 'backup'):
                print(f'too small for {XMAS} in {k} direction!')
                continue
            if nrows-row < NCHRS-1 and k in {"down", 'frontdown', 'backdown'}:
                print(f'too small for {XMAS} in {k} direction!')
                continue
            print(f'is {xchr} in {k} direction?')
            if wordsearch[v[0]][v[1]] == xchr:
                print('yep!')
                try:
                    retv = test_all_around(wordsearch, v[0], v[1], xchr=XMAS[2], direction=k)
                except Exception as exc:
                    print(exc)
                    print('... trying a different direction -->')
                    continue
                else:
                    print('success!')
                    retv = retv[0], (row, col), retv[1]
                    break
            else:
                print('nope!')
    else:
        k, v = direction, test_locs[direction]
        print(f'searching in row {v[0]}, col {v[1]} for {xchr} in {direction} direction')
        if wordsearch[v[0]][v[1]] == xchr:
            print('yep!')
            if xchr == "S":
                return k, (v[0], v[1])
            retv = test_all_around(wordsearch, v[0], v[1], xchr=XMAS[3], direction=k)
        else:
            raise Exception(f'{xchr} not found at ({v[0]}, {v[1]})')
    return retv

test_all_around(test_data, 0, col=x[0][0])
```
